```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('brewSales.csv')
data.describe()
```

```python
beerSample = data
```

```python
beerSample.head()
```

```python
beerSample['date'] = pd.to_datetime(beerSample['date'])
draughtSales = beerSample[beerSample['saleType'] == 'draught']
bottleSales = beerSample[beerSample['saleType'] == 'bottles or cans']
bottleSales.head()
```

```python
draughtSales.describe()
```

```python
bottleSales.head()
```

```python
beerSample = beerSample.groupby(['date']).sum()
draughtSales = draughtSales.groupby(['date']).sum()
bottleSales = bottleSales.groupby(['date']).sum()

# draughtSales.head()

# beerSample = beerSample.set_index('date')
# draughtSales = draughtSales.set_index('date')
# bottleSales = bottleSales.set_index('date')
```

```python
def multiplySales (row):
    return row['purchase'] * 10

beerSample['totalSales'] = beerSample.apply(lambda row: multiplySales(row), axis=1)
draughtSales['totalSales'] = draughtSales.apply(lambda row: multiplySales(row), axis=1)
bottleSales['totalSales'] = bottleSales.apply(lambda row: multiplySales(row), axis=1)
```

```python
draughtSales.head()
```

```python
beerSample = beerSample.sort_index()
draughtSales = draughtSales.sort_index()
bottleSales = bottleSales.sort_index()
```
